```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df
```

```python
test_df
```

```python
# Convert categorical data to numeric and separate target feature for training data
y_train = pd.DataFrame(train_df['loan_status'])
X_train= train_df.drop(columns=['loan_status', 'Unnamed: 0', 'index'])
X_train.head()
```

```python
X_train.shape, y_train.shape
```

```python
# Convert categorical data to numeric and separate target feature for testing data
y_test = pd.DataFrame(test_df['loan_status'])
X_test = test_df.drop(columns=['loan_status', 'Unnamed: 0', 'index'])
X_test.head()
```

```python
set(train_df["loan_status"])
```

```python
set(test_df["loan_status"])
```

```python
# Convert categorical data to numeric and separate target label for training data
# use pd.get_dummies to convert categorial data to numeric
X_train = pd.get_dummies(X_train)
```

```python
X_test = pd.get_dummies(X_test)
```

```python
# dummies_train_df.columns.shape
```

```python
# dummies_test_df.columns.shape
```

```python
#see the difference between the two dummies df

set(X_train.columns) - set(X_test.columns)

```

```python
# X_train["debt_settlement_flag_Y"]
# X_test["debt_settlement_flag_Y"]
set(X_test.columns) - set(X_train.columns)
```

```python
# add missing dummy variables to testing set

X_test['debt_settlement_flag_Y'] = 0





```

```python
#check again if X_Train has more columns then X_Test:
set(X_train.columns) - set(X_test.columns)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
#Random Forest Model

# X is the input values, y is the output values
model = RandomForestClassifier()
model.fit(X_train,y_train)

#insert the X_test values to get a prediction
predictions = model.predict(X_test)

#to see how accurate the model is, we simply compare the "predictions for output test values", with the "y_test values".
score = accuracy_score(y_test, predictions)
score
```

```python
#Logistic Regression Classifier

# X is the input values, y is the output values
model = LogisticRegression()
model.fit(X_train,y_train)

#insert the X_test values to get a prediction
predictions = model.predict(X_test)

#to see how accurate the model is, we simply compare the "predictions for output test values", with the "y_test values".
score = accuracy_score(y_test, predictions)
score

```

```python
# standardize the features with a scaler

scaler = StandardScaler()
scaler.fit(X_train)

# transform the data, fit it into the new space

X_train_scaled = scaler.transform(X_train)
X_train_scaled.shape


```

```python
X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
#Random Forest Model (scaled)

# X is the input values, y is the output values
model = RandomForestClassifier()
model.fit(X_train_scaled,y_train)

#insert the X_test values to get a prediction
predictions = model.predict(X_test_scaled)

#to see how accurate the model is, we simply compare the "predictions for output test values", with the "y_test values".
score = accuracy_score(y_test, predictions)
score
```

```python
#Logistic Regression Classifier (Scaled)

# X is the input values, y is the output values
model = LogisticRegression()
model.fit(X_train_scaled,y_train)

#insert the X_test values to get a prediction
predictions = model.predict(X_test_scaled)

#to see how accurate the model is, we simply compare the "predictions for output test values", with the "y_test values".
score = accuracy_score(y_test, predictions)
score
```

```python
Summary:
    
Prediction Before standerdizing:
    I belive the logistic regression model wouldn't work well for our data, it's because it needs normal distribution data. So the Random forest model would work better here.
    
Prediction After standerdizing:
    I belive the logistic regression model will work better, since standardizing is making the data into normal distribution, which works well with logistic model.

Final Summary:
    We observe that the Random FOrest model score stays around 0.63~0.64, but the score of Logistic Model score improves alot from
0.5168013611229264 to 0.767333049766057.  So it improves by having normal distribution data.

    
    
    
    
    
```
